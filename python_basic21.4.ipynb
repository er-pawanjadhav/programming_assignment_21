```python
def compound():
    p = float(input("enter initial amount : "))
    r =  float(input("enter rate of intrest : "))
    t =  float(input("enter time period : "))
    n = float(input("enter no of months : "))
    t = n*t
    r = r/n
    n = r+1
    n = n**t
    p = p*n
    return p

print(f"compund intrest is {compound()}")
```
